```python
bronze_path ="abfss://bronze@pharmasalesdatalake.dfs.core.windows.net/openfda/drug"
```

```python
df_raw = spark.read.json(bronze_path)
display(df_raw)
```

```python
from pyspark.sql.functions import col, explode
```

```python
df_exploded = df_raw.select(explode("results").alias("r"))
```

```python
display(df_exploded.limit(10))
```

```python
df_flattened = df_exploded.select(
    col("r.product_ndc").alias("product_ndc"),
    col("r.generic_name").alias("generic_name"),
    col("r.brand_name").alias("brand_name"),
    col("r.labeler_name").alias("labeler_name"),
    col("r.dosage_form").alias("dosage_form"),
    col("r.route").alias("route_name"),
    col("r.product_type").alias("product_type"),
    col("r.marketing_category").alias("marketing_category")

)
```

```python
display(df_flattened.limit(10))
```

```python
from pyspark.sql.functions import concat_ws, regexp_replace

df_clean1 = df_flattened \
    .withColumn("route_name", concat_ws(",", "route_name")) \
    .withColumn("route_name", regexp_replace("route_name", '\[|\]|"', ""))
display(df_clean1.limit(10))
```

```python
from pyspark.sql.functions import trim,nullif

df_clean = df_clean1.dropDuplicates().na.drop(subset=["product_ndc"]) \
    .withColumn("brand_name", trim(col("brand_name"))) \
    .withColumn("generic_name", trim(col("generic_name"))) \
    .withColumn("labeler_name", trim(col("labeler_name")))


display(df_clean.limit(10))
```

```python
silver_path = "abfss://silver@pharmasalesdatalake.dfs.core.windows.net/openfda/drug/"
df_clean.write.format("delta").mode("overwrite").save(silver_path)
```
